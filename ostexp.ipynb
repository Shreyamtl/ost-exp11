```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['data.csv']))
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
print("Average Age:", df['Age'].mean())
print("Average Salary:", df['Salary'].mean())
```

```python
df['Age'].hist(bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Employees')
plt.show()
```

```python
plt.bar(df['Name'], df['Salary'], color='purple', edgecolor='black')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Bar Plot: Name vs Salary')

plt.show()
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter plot ')
plt.show()
```

```python
plt.plot(df['Age'], df['Salary'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Line Plot: Age vs Salary')
plt.show()
```

```python
plt.boxplot(df['Salary'])
plt.ylabel('Salary')
plt.title('Box Plot of Salary')
plt.show()
```

```python
sns.relplot(x='Age', y='Salary', data=df, kind='scatter', height=5, aspect=1.3)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Relational Plot ')
plt.show()
```

```python
sns.pairplot(df[['Name','Age', 'Salary']])
plt.suptitle('Pair Plot of Age and Salary', y=1.02)
plt.show()
```

```python
sns.jointplot(x='Age', y='Salary', data=df, kind='scatter', color='blue')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Age','Salary']].corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
